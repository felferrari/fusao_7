```python
import numpy as np
import os
import matplotlib.pyplot as plt
```

## Transformer

```python
exp_name = 'tr_sm_opt_pm_nc_5'
exp_path = os.path.join('D:', 'Ferrari', 'exps_7', exp_name)
visual_path = os.path.join(exp_path, 'visual')
```

```python
curve_tr = np.load(os.path.join(visual_path, f'curve_0.npy'))
prec_tr = curve_tr[:,0]
recall_tr = curve_tr[:,1]

recall_tr_ = np.insert(recall_tr, 0, 0)
prec_tr_ = np.insert(prec_tr, 0, prec_tr[0])
deltaR = recall_tr_[1:]-recall_tr_[:-1]
m_prec_tr_ = (prec_tr_[1:] + prec_tr_[:-1])/2
AP_tr = np.sum(m_prec_tr_*deltaR)
```

## ResUNET

```python
exp_name = 'rs_sm_opt_pm_nc_3'
exp_path = os.path.join('D:', 'Ferrari', 'exps_7', exp_name)
visual_path = os.path.join(exp_path, 'visual')
```

```python
curve_rs = np.load(os.path.join(visual_path, f'curve_0.npy'))
prec_rs = curve_rs[:,0]
recall_rs = curve_rs[:,1]

recall_rs_ = np.insert(recall_rs, 0, 0)
prec_rs_ = np.insert(prec_rs, 0, prec_rs[0])
deltaR = recall_rs_[1:]-recall_rs_[:-1]
m_prec_rs_ = (prec_rs_[1:] + prec_rs_[:-1])/2
AP_rs = np.sum(m_prec_rs_*deltaR)
```

## Graphic

```python
plt.figure(figsize=(10,10))
plt.plot(recall_tr_, prec_tr_, 'b-', label = f'Transformer AP: {AP_tr:.4f}')
plt.plot(recall_rs_, prec_rs_, 'r-', label = f'ResUNET AP: {AP_rs:.4f}')
plt.legend(loc="lower left")
ax = plt.gca()
ax.set_ylim([0,1.01])
ax.set_xlim([0,1.01])
plt.grid()
```
